```python
#1.
def clean_and_count(text):
    cleaned = text.strip().lower()
    return len(cleaned)

# Sample
text = "   Hello World  "
print(clean_and_count(text))   # Expected: 11

#2.
def filter_even(nums):
    return [x for x in nums if x % 2 == 0]

# Sample
nums = [1, 2, 3, 4, 5, 6]
print(filter_even(nums))   # Expected: [2, 4, 6]

#3.
def word_frequency(words):
    freq = {}
    for w in set(words):
        freq[w] = words.count(w)
    return freq

# Sample
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
print(word_frequency(words))

#4.
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero error"

# Sample
print(safe_divide(10, 2))
print(safe_divide(5, 0))

#5.
def calculate_grade(marks):
    avg = sum(marks) / len(marks)

    if avg >= 80:
        return "A"
    elif 60 <= avg <= 79:
        return "B"
    elif 40 <= avg <= 59:
        return "C"
    else:
        return "F"

# Sample
print(calculate_grade([90, 80, 70]))  # A
print(calculate_grade([50, 45, 40]))  # C
print(calculate_grade([30, 35]))      # F

#6.
nums = [-3, -1, 0, 2, 5]
positive_nums = list(filter(lambda x: x > 0, nums))
print(positive_nums)   # Expected: [2, 5]

#7.
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance


# Sample
acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.get_balance())   # 1300

acc.withdraw(2000)
print(acc.get_balance())   # 1300

#8.
def filter_pune_young(df):
    return df[(df["city"] == "Pune") & (df["age"] < 25)]

# Sample DataFrame
data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [24, 30, 22, 28],
    "city": ["Pune", "Mumbai", "Pune", "Delhi"]
}
df = pd.DataFrame(data)

result = filter_pune_young(df)
print(result)

#9.
temps = [29, 31, 30, 32, 33, 35, 34]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt.plot(days, temps, marker='o')
plt.title("Day Wise Temperature Variation")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()

#10.
sales = np.array([1200, 1500, 1000, 1300, 1700, 1400, 1600])

total_revenue = sales.sum()

daily_percentage = np.round((sales / total_revenue) * 100, 2)

print("Total Revenue:", total_revenue)
print("Daily Percentage:", daily_percentage)





# =============================== PRACTICE QUESTION 1 ===============================
1.	Write a python code that takes two numbers as input and performs addition, subtraction, multiplication, and division. Print the results. 
2.	Write a python code that takes two numbers and returns the larger of the two without using functions. 
3.	Write a python code to merge two lists and get all the elements in a single list. 
4.	Write a python code to define a nested Lists (2D List) and show how to access its elements. 
5.	Input a list of 10 elements and try to add 3 more elements in the last using python code. 
6.	Write a python code to get 5 elements from user for forming a list. Also print the final and reverse list. 
7.	Find the largest and smallest number in the list which taken as input from user using list operations with python code. 
8.	Given a dictionary of students and their favourite colours: 
people={'Arham':'Blue','Lisa':'Yellow',''Vinod:'Purple','Jenny':'Pink'} a) Find out how many students are in the list 
b)	Change Lisa’s favourite colour 
c)	Remove 'Jenny' and her favourite colour 
d)	Sort and print students and their favourite colours alphabetically by name 
9.	Write a Python program to swap two variables. 
10.	Write a python program to swap two variables without using a third variable. 
11.	Write a Python program to find the repeated items of a tuple. 
12.	Write a Python program to create multiple lists. 
13.	Write a python code for creating 2 sets then perform union intersection and difference operations on them. 
14.	write a python code to input a float variable from user then convert it into integer type. 

    
# 1.1.	Write a python code that takes two numbers as input and performs addition, subtraction, multiplication, and division. Print the results. 
a = float(input("Enter number: "))   # take first number
b = float(input("Enter number: "))   # take second number
print(a + b, a - b, a * b, a / b)    # print results of operations

# 2.2.	Write a python code that takes two numbers and returns the larger of the two without using functions. 
a = int(input("Enter number: "))     # take first integer
b = int(input("Enter number: "))     # take second integer
print(a if a > b else b)             # print the larger number

# 3. Merge two lists.
l1 = [1, 2, 3]                       # first list
l2 = [4, 5, 6]                       # second list
print(l1 + l2)                       # merged list

# 4. Nested list access.
lst = [[1,2,3],[4,5,6]]              # 2D list
print(lst[0][1])                     # access 2nd element of 1st row

# 5. Add 3 more elements to existing 10-element list.
lst = list(range(10))                # list of 10 elements
lst.extend([11, 12, 13])             # add 3 elements
print(lst)                           # print final list

# 6. Input 5 elements & print list + reverse.
lst = [input("Enter item: ") for _ in range(5)]  # take 5 inputs
print(lst)                                       # print list
print(lst[::-1])                                 # print reverse list

# 7. Largest & smallest in list.
lst = list(map(int, input("Enter numbers: ").split())) # input numbers
print(max(lst), min(lst))                              # print max/min

# 8. Dictionary operations.
people = {'Arham':'Blue','Lisa':'Yellow','Vinod':'Purple','Jenny':'Pink'}  # dictionary
print(len(people))                                   # count entries
people['Lisa'] = 'Black'                             # change Lisa's value
people.pop('Jenny')                                  # remove Jenny
print(dict(sorted(people.items())))                  # print sorted dictionary

# 9. Swap two variables.
a, b = 5, 10                     # original values
a, b = b, a                      # swap
print(a, b)                      # print swapped values

# 10. Swap without third variable.
a, b = 3, 7                      # original numbers
a = a + b                        # store sum
b = a - b                        # compute new b
a = a - b                        # compute new a
print(a, b)                      # print values

# 11. Repeated items in tuple.
t = (1,2,3,2,4,1)                # tuple
print([x for x in set(t) if t.count(x) > 1])  # identify repeating items

# 12. Create multiple lists.
l1 = [1,2]                       # list 1
l2 = [3,4]                       # list 2
l3 = [5,6]                       # list 3

# 13. Set operations.
s1 = {1,2,3}                     # first set
s2 = {3,4,5}                     # second set
print(s1 | s2, s1 & s2, s1 - s2)  # union, intersection, difference

# 14. Convert float to int.
x = float(input("Enter float: "))  # input float
print(int(x))                       # print integer conversion



# =============================== PRACTICE QUESTION 2 ===============================

Assignment 2 
1.	Write a Python program to do arithmetical operations addition, subtraction, multiplication and division. 
2.	Write a Python program to find the area of a triangle. 
3.	Write a program to display calendar of any month using calendar module of python. 
4.	Write a Python Program to Check if a Number is Positive, Negative or Zero. 
5.	Write a Python Program to Check if a Number is Odd or Even. 
6.	Write a Python Program to Check Leap Year. The year should be entered by user. 7. Write a Python Program to Print all Prime Numbers in an Interval of 1-10 
8.	Write a Python Program to Find the Factorial of a Number. 
9.	Write a Python Program to Display the multiplication Table of any number entered by user. 
10.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations. 
11.	Write a Python Program to calculate Body Mass Index of a person. Also print if the person is underweight, overweight or obese. 
12.	Write a Python Program to find sum of array of 5 elements. 
13.	Write a Python Program to find largest and smallest element in an array. 
14.	Write a Python program to Multiply all numbers in the list 7 elements. 

    
# 1. Arithmetical operations.
a = float(input("Enter number: "))  # input num1
b = float(input("Enter number: "))  # input num2
print(a+b, a-b, a*b, a/b)           # print results

# 2. Area of triangle.
b = float(input("Enter base: "))    # base
h = float(input("Enter height: "))  # height
print(0.5 * b * h)                  # area formula

# 3. Calendar of any month.
import calendar                     # import calendar module
y = int(input("Enter year: "))      # year input
m = int(input("Enter month: "))     # month input
print(calendar.month(y, m))         # print calendar

# 4. Positive/Negative/Zero.
n = int(input("Enter number: "))    # input number
print("Positive" if n > 0 else "Negative" if n < 0 else "Zero")  # check value

# 5. Odd or even.
n = int(input("Enter number: "))    # number input
print("Even" if n % 2 == 0 else "Odd")  # check even/odd

# 6. Leap year.
y = int(input("Enter year: "))      # year
print("Leap Year" if (y%4==0 and y%100!=0) or y%400==0 else "Not Leap") # condition

# 7. Prime numbers 1–10.
for n in range(2, 11):              # loop 2–10
    for i in range(2, n):           # divisor check
        if n % i == 0: break        # not prime
    else:
        print(n)                    # prime number

# 8. Factorial.
n = int(input("Enter number: "))    # number input
fact = 1                            # start factorial
for i in range(1, n+1): fact *= i   # multiply repeatedly
print(fact)                         # print result

# 9. Multiplication table.
n = int(input("Enter number: "))    # input number
for i in range(1, 11):              # loop 1–10
    print(n, "x", i, "=", n*i)      # print table

# 10. Simple calculator.
a = float(input("Enter number: "))  # first number
b = float(input("Enter number: "))  # second number
print(a+b, a-b, a*b, a/b)           # results

# 11. BMI calculator.
w = float(input("Enter weight: "))  # weight
h = float(input("Enter height: "))  # height
print(w / (h*h))                    # BMI formula

# 12. Sum of array of 5 elements.
arr = [int(input("Enter element: ")) for _ in range(5)]  # 5 elements
print(sum(arr))                         # print sum

# 13. Largest & smallest element.
arr = list(map(int, input("Enter values: ").split()))  # input list
print(max(arr), min(arr))              # max and min

# 14. Multiply all list numbers.
lst = [1,2,3,4,5,6,7]                  # fixed list
p = 1                                  # product start
for x in lst: p *= x                   # multiply each element
print(p)                               # print product



# =============================== PRACTICE QUESTION 3 ===============================
Assignment 3 
1.	Write a program that takes a numeric grade (0-100) as input and outputs a letter grade based on the following criteria: 
•	A: 90-100 
•	B: 80-89 
•	C: 70-79 
•	D: 60-69 
•	F: below 60 
2.	Write a program that calculates the Body Mass Index (BMI) based on user input for weight (in kg) and height (in meters). Then classify the result as: 
•	"Underweight" if BMI < 18.5 
•	"Normal weight" if 18.5 ≤ BMI < 24.9 
•	"Overweight" if 25 ≤ BMI < 29.9 
•	"Obesity" if BMI ≥ 30 
3.	Develop a program that takes a year as input and determines if it is a leap year. 
4.	Write a program that uses a for loop to print the numbers from 1 to 10, each on a new line. 
5.	Create a program that uses a while loop to calculate the sum of all integers from 1 to a given number (inclusive) that the user inputs. 
6.	Write a program that prints all even numbers from 1 to 20 using a for loop. 
7.	Develop a program that takes a non-negative integer as input and calculates its factorial using a for loop. 
8.	Create a program that prints the multiplication table for a given number (1 to 10) using a for loop. 
9.	Create a program that takes a string input from the user and prints it in reverse using a 
for loop. 
10.	Write a program that takes a positive integer as input and calculates the sum of its digits using a while loop. 

# 1. Grade system.
n = int(input("Enter grade: "))        # numeric grade
if n>=90: print("A")                   # grade A
elif n>=80: print("B")                 # grade B
elif n>=70: print("C")                 # grade C
elif n>=60: print("D")                 # grade D
else: print("F")                       # grade F

# 2. BMI categories.
w = float(input("Enter weight: "))     # weight
h = float(input("Enter height: "))     # height
b = w/(h*h)                            # BMI calculation
if b<18.5: print("Underweight")        # category
elif b<24.9: print("Normal weight")
elif b<29.9: print("Overweight")
else: print("Obesity")

# 3. Leap year check.
y = int(input("Enter year: "))         # year input
print("Leap" if (y%4==0 and y%100!=0) or y%400==0 else "Not Leap") # condition

# 4. Print numbers 1–10.
for i in range(1, 11): print(i)        # loop

# 5. Sum 1 to n using while loop.
n = int(input("Enter number: "))       # input number
s = 0                                  # sum start
i = 1                                  # counter
while i <= n:                          # loop condition
    s += i                             # add
    i += 1                             # increment
print(s)                               # print sum

# 6. Print even numbers 1–20.
for i in range(2, 21, 2): print(i)     # step 2 to get even numbers

# 7. Factorial using loop.
n = int(input("Enter number: "))
f = 1                                   # factorial start
for i in range(1, n+1): f *= i          # multiply
print(f)                                # result

# 8. Multiplication table 1–10.
n = int(input("Enter number: "))
for i in range(1, 11): print(n*i)       # print multiples

# 9. Reverse string using a loop.
s = input("Enter string: ")
rev = ""                                # empty string
for ch in s: rev = ch + rev             # prepend character
print(rev)                              # reversed string

# 10. Sum of digits using while.
n = input("Enter number: ")             # string input
s = 0                                   # sum start
i = 0                                   # index
while i < len(n):                       # loop through digits
    s += int(n[i])                      # convert char to int
    i += 1                              # next position
print(s)                                # print result



# =============================== PRACTICE QUESTION 4 ===============================
Assignment 4 
1.	Write a Python program to split and join a string. 
2.	Write a Python program to find the sum of all items in a dictionary. 
3.	Write a Python program to Merging two Dictionaries. 
4.	Write a Python program to sort Python Dictionaries by Key or Value. Sample 
Dictionary = {'apple': 3, 'banana': 1, 'cherry': 2, 'date': 4} 
5.	Write a python code to calculate square root of a number entered by user. 
6.	Define a class Person and its two child classes: Male and Female. All classes have a method "getGender" which can print "Male" for Male class and "Female" for Female class. 
7.	Define a class named Shape and its subclass Square. The Square class has an init function which takes a length as argument. Both classes have an area function which can print the area of the shape where Shape's area is 0 by default. 
8.	Write a function that calculates the factorial of a number recursively. Take number from user. 
9.	Create the function that takes a list of dictionaries and returns the sum of people's budgets. Examples  
get_budgets([ { 'name': 'John', 'age': 21, 'budget': 23000 }, { 'name': 'Steve', 'age': 32, 'budget': 40000 }, { 'name': 'Martin', 'age': 16, 'budget': 2700 } ]) ➞ 65700 get_budgets([ { 'name': 'John', 'age': 21, 'budget': 29000 }, { 'name': 'Steve', 'age': 32, 
'budget': 32000 }, { 'name': 'Martin', 'age': 16, 'budget': 1600 } ]) ➞ 62600 
10.	Write a class named Rectangle that has methods to calculate its area and perimeter. 

# 1. Split & join string.
s = "hello world"                       # sample string
print(" ".join(s.split()))              # split → join

# 2. Sum of dictionary items.
d = {'a':10,'b':20}                     # dictionary
print(sum(d.values()))                  # sum values

# 3. Merge dictionaries.
d1 = {'a':1}                            # first dict
d2 = {'b':2}                            # second dict
d1.update(d2)                           # merge
print(d1)                               # print result

# 4. Sort dictionary by key or value.
d = {'apple':3,'banana':1,'cherry':2,'date':4}
print(dict(sorted(d.items())))          # sort by key
print(dict(sorted(d.items(), key=lambda x:x[1])))  # sort by value

# 5. Square root of number.
import math                             # import math
n = float(input("Enter number: "))      # number
print(math.sqrt(n))                     # sqrt

# 6. Person, Male, Female classes.
class Person:
    def getGender(self): return "Unknown"  # base class

class Male(Person):
    def getGender(self): return "Male"     # override

class Female(Person):
    def getGender(self): return "Female"   # override

# 7. Shape & Square class.
class Shape:
    def area(self): return 0               # default

class Square(Shape):
    def __init__(self,l): self.l = l       # store length
    def area(self): return self.l * self.l # area

# 8. Recursive factorial.
def fact(n):                               # define function
    return 1 if n==0 else n * fact(n-1)    # recursion
print(fact(int(input("Enter number: "))))  # call function

# 9. Sum of budgets.
def get_budgets(lst):                      # define function
    return sum(p['budget'] for p in lst)   # sum budgets

# 10. Rectangle class.
class Rectangle:
    def __init__(self,w,h):                # constructor
        self.w = w; self.h = h
    def area(self): return self.w*self.h   # area formula
    def perimeter(self): return 2*(self.w+self.h) # perimeter



# =============================== PRACTICE QUESTION 5 ===============================
Assignment 5 
 
1.	Create a Vehicle class with attributes number of wheels,
number of airbags, engine horsepower, max speed, mileage, etc. 
Create a constructor and a method that displays vehicle info. 
 
2.	Create a child class Bus, from Vehicle class,
that will inherit all of the variables and methods of the Vehicle class.
This Bus class will have its own attributes – number of doors, number of seats,
fare per km, etc, along with inherited attributes. Write a method to calculate bus
fare by accepting distance in km as parameter. 
 
 
3.Create class BankAccount with attributes cutomer_id, customer_name, 
account_balance.Write methods to withdraw and deposit amounts. 
If during withdrawing money from the account, the account balance is not
sufficient, display appropriate messages. Create objects of this class and
test methods to withdraw and deposit. 
 
4.Create class RegularAccount derived from base class BankAccount.
There is a penalty of Rs. 10.00 if the balance falls below a minimum of Rs.3000. 
If customer tries to withdraw amount and balance goes below Rs. 3000,
deduct penalty amount from the balance and notify customer about it, 
and if after deducting penalty amount if balance goes in below zero notify customer about insufficient balance. 
 
 
5.Create a class hierarchy with a base class Shape and two subclasses Circle and Square. 
The Shape class should have a method area() that returns 0 (as a default).
The Circle class should override the area() method to calculate the area of a circle using the formula π * radius^2.
The Square class should override the area() method to calculate the area of a square using the formula side^2. 
Create a function print_area(shape) that takes an object of the Shape class (or its subclasses) and prints its area. 
 
6.Imagine you are building a simulation for different types of employees in a company. 
Create a base class Employee and two subclasses Manager and Developer.
Each subclass should implement a method salary() that returns the monthly salary of the employee, but with different logic for managers and developers. 
 

# 1. Vehicle class.
class Vehicle:
    def __init__(self, wheels, airbags, hp, max_speed, mileage):
        self.wheels = wheels              # number of wheels
        self.airbags = airbags            # number of airbags
        self.hp = hp                      # horsepower
        self.max_speed = max_speed        # max speed
        self.mileage = mileage            # mileage
    def display_info(self):
        print(self.wheels, self.airbags, self.hp, self.max_speed, self.mileage)  # print info

# 2. Bus class.
class Bus(Vehicle):
    def __init__(self, wheels, airbags, hp, max_speed, mileage, doors, seats, fare):
        super().__init__(wheels, airbags, hp, max_speed, mileage) # inherit Vehicle
        self.doors = doors               # number of doors
        self.seats = seats               # seat count
        self.fare = fare                 # fare per km
    def calculate_fare(self, km):
        return km * self.fare            # fare = distance × rate

# 3. BankAccount class.
class BankAccount:
    def __init__(self, id, name, balance):
        self.id = id                     # customer ID
        self.name = name                 # customer name
        self.balance = balance           # current balance
    def deposit(self, amt):
        self.balance += amt              # add money
        print(self.balance)              # print balance
    def withdraw(self, amt):
        if amt > self.balance:           # insufficient funds check
            print("Insufficient Balance")
        else:
            self.balance -= amt          # deduct money
            print(self.balance)          # print balance

# 4. RegularAccount class.
class RegularAccount(BankAccount):
    def withdraw(self, amt):
        if self.balance - amt < 3000:    # below minimum balance
            self.balance -= 10           # penalty
            if self.balance - amt < 0:   # still insufficient
                print("Insufficient Balance")
                return
        self.balance -= amt              # deduct amount
        print(self.balance)              # print new balance

# 5. Shape → Circle & Square classes.
class Shape:
    def area(self): return 0             # default area

class Circle(Shape):
    def __init__(self, r): self.r = r    # radius
    def area(self): return 3.14159 * self.r * self.r  # circle area

class Square(Shape):
    def __init__(self, s): self.s = s    # side
    def area(self): return self.s * self.s  # square area

def print_area(shape): print(shape.area())   # print area

# 6. Employee → Manager & Developer salary system.
class Employee:
    def salary(self): return 0           # default

class Manager(Employee):
    def __init__(self, base): self.base = base  # base salary
    def salary(self): return self.base + 5000   # manager bonus

class Developer(Employee):
    def __init__(self, base): self.base = base  # base salary
    def salary(self): return self.base + 2000   # developer bonus

```
